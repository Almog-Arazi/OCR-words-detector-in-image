build dilate and erode functions


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

figsize = (10, 10)

```

```python

img = np.zeros((50, 50))
img[20:30, 20:30] = 1

plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray")
plt.show()

```

```python
kernel = np.zeros((5,5),dtype=np.uint8)
kernel[2,:] = 1
kernel[:,2] = 1


plt.figure(figsize=figsize)
plt.imshow(kernel,cmap="gray")
plt.show()

```

```python
def my_dilate(img,kernel):
    
    img_height = img.shape [0]
    img_width = img.shape [1]

    kernel_height = kernel.shape[0]
    kernel_width = kernel.shape[1]
 
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2

    padded_img = np.pad(img, ((pad_height, pad_height), (pad_width, pad_width)))

    res_img = np.zeros((50, 50))

    for i in range(img_height): # each row
        for j in range(img_width): # each column
            region = padded_img[i:i+kernel_height, j:j+kernel_width]  
            if np.sum(region * kernel) >= 1:
                res_img[i, j] = 1

    return res_img

plt.figure(figsize=figsize)
plt.imshow(my_dilate(img,kernel),cmap="gray")
plt.title("my_dilate result Image")
plt.show()

```

```python

difference = np.abs(cv2.dilate(img, kernel) -  my_dilate(img, kernel))

if np.sum(difference) == 0:
        print("cv2.dilate & my_dilate are the same!")
else:
        print("try again...")


```

```python
def my_erode(img,kernel):
  
   
    img_height = img.shape [0]
    img_width = img.shape [1]

    kernel_height = kernel.shape[0]
    kernel_width = kernel.shape[1]
 
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2

    padded_img = np.pad(img, ((pad_height, pad_height), (pad_width, pad_width)))

    res_img = np.zeros((50, 50))

    for i in range(img_height): # each row
        for j in range(img_width): # each row
            region = padded_img[i:i+kernel_height, j:j+kernel_width]  
            
            # Apply the dilation condition
            if np.sum(region * kernel) >= np.sum(kernel):
                res_img[i, j] = 1

    return res_img

plt.figure(figsize=figsize)
plt.imshow(my_erode(img,kernel),cmap="gray")
plt.title("my_erode result Image")
plt.show()
```

```python

difference = np.abs(cv2.erode(img, kernel) -  my_erode(img, kernel))

if  np.sum(difference) == 0:
    print("cv2.erode & my_erode are the same!")
else: 
    print("try again...")

```
